```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

platforms = ["Netflix", "YouTube"]
age_groups = ["<18", "18-24", "25-34", "35-44", "45+"]
genres = ["Movies", "TV Shows", "Music", "Gaming", "Sports", "Education"]
devices = ["Mobile", "Laptop", "TV", "Tablet"]

data = {
    "Platform": np.random.choice(platforms, 300),
    "Watch_Hours": np.random.randint(1, 20, 300),
    "Age_Group": np.random.choice(age_groups, 300),
    "Genre": np.random.choice(genres, 300),
    "Device": np.random.choice(devices, 300)
}

df= pd.DataFrame(data)

print(df.head())

platform_age = df.groupby(["Platform", "Age_Group"])["Watch_Hours"].mean().reset_index()
print("\nAverage Watch Hours by Platform & Age Group:\n", platform_age)

avg_watch = df.groupby("Platform")["Watch_Hours"].mean()
avg_watch.plot(kind="bar", color=["red", "blue"], alpha=0.7)
plt.title("Average Watch Hours: Netflix vs YouTube")
plt.ylabel("Average Hours")
plt.show()



```

```python
pivot_age = platform_age.pivot(index="Age_Group", columns="Platform", values="Watch_Hours")
pivot_age.plot(kind="bar", figsize =(8,5))
plt.title("Average Watch Hours by Age Group")
plt.ylabel("Average Hours")
plt.show()
```

```python
genre_platform = df.groupby(["Genre", "Platform"])["Watch_Hours"].mean().unstack()
genre_platform.plot(kind="bar", figsize=(10,6))
plt.title("Average Watch Hours by Genre (Netflix vs YouTube)")
plt.ylabel("Average Hours")
plt.show()
```
